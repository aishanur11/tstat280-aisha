#### Exercise 1

Create a connection to the bike database found in `data/bike_store.db`. Then create a cursor that can be used to perform queries on this database.

```python
# Put your code here
import sqlite3
from pprint import pprint
conn = sqlite3.connect("../data/bike_store.db")
cur = conn.cursor()
```

#### Exercise 2

Write a SQL query to display the first 5 customers from the `customers` table.

Include the following columns:
- customer_id
- first_name
- last_name

```python
# query the first 5 customers
res = cur.execute ("""
    SELECT customer_id,first_name,last_name
    FROM customers
    LIMIT 5
    
""")
#display the result
res.fetchall()
```

Write a SQL query to calculate the following from the `products` table:

- The total number of products
- The average list price
- The highest list price
- The lowest list price

```python

res = cur.execute("""
    SELECT COUNT(*), AVG(list_price), MAX(list_price), MIN(list_price) FROM 
    products    
""")
res.fetchall()
```

#### Exercise 4

Write a SQL query to retrieve customers from the `customers` table who meet ALL of the following conditions:

- They live in either 'CA' or 'NY'
- Their first name starts with the letter 'S'
- AND (they are in the city 'Los Angeles' OR 'New York')

Include the following columns:
- customer_id
- first_name
- last_name
- city
- state

```python
res = cur.execute("""
    SELECT customer_id, first_name, last_name, city, state
    FROM customers 
    WHERE state IN ("CA", "NY") AND 
    first_name LIKE "S%" AND 
    city IN ("Los Angeles", "New York")
    
""")
res.fetchall()
```

#### Exercise 5

Write a SQL query using the `orders` table to find the total number of orders placed by each customer.

- Only include customers who have placed 2 or more orders
- Sort the results by total number of orders in descending order

Include the following columns:
- customer_id
- total_orders (use COUNT)

```python
# Put your code here
res = cur.execute ("""
    SELECT customer_id, COUNT(order_id) AS total_orders 
    FROM orders 
    GROUP BY customer_id
    HAVING COUNT(order_id) >= 2
    ORDER BY total_orders DESC
    
""")
res.fetchall()
```

#### Exercise 6

Write a SQL query to display all customers and any orders they have placed.

- Include customers even if they have not placed any orders

Include the following columns:
- customer_id
- first_name
- last_name
- order_id


```python
# Put your code here
res =  cur.execute ("""
    SELECT c.customer_id, first_name, last_name, order_id 
    FROM customers c LEFT JOIN  orders o
    ON c.customer_id = o.customer_id

    
""")
res.fetchall()
```
