Create a connection to the bike database found in `data/bike_store.db`. Then create a cursor that can be used to perform queries on this database.

#### Exercise 2

Write a SQL query to display the first 5 customers from the `customers` table.

Include the following columns:
- customer_id
- first_name
- last_name

```python
# query the first 5 customers
res = cur.execute ("""
    SELECT customer_id,first_name,last_name
    FROM customers
    LIMIT 5
    
""")
#display the result
res.fetchall()
```

#### Exercise 3

Write a SQL query to calculate the following from the `products` table:

- The total number of products
- The average list price
- The highest list price
- The lowest list price

```python
# query the total number of products
rescount = cur.execute("""
    SELECT COUNT(*) FROM 
    products    
""")
rescount.fetchall()
```

```python
# query avg list price
resavg = cur.execute("""
    SELECT AVG(list_price) FROM
    products
""")
resavg.fetchall()
```

```python
#highest list price
ressum = cur.execute("""
    SELECT SUM(list_price) FROM
    products
""")
ressum.fetchall()
```

```python
#highest list price
resmax = cur.execute("""
    SELECT MAX(list_price) FROM
    products
""")
resmax.fetchall()
```

```python

#lowest list price
resmin = cur.execute("""
    SELECT MIN(list_price) FROM
    products
""")
resmin.fetchall()
```
