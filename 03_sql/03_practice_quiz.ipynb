Write a single SQL query to analyze customer order activity.

- Join `customers` and `orders` using a LEFT JOIN
- Use the relationship between primary and foreign keys (customer_id)

Filter the data so that:
- Customers are in states 'TX', 'CA', or 'NY'
- AND their last name contains the letter 's'
- OR they are in the city 'Seattle'

For each customer, calculate:
- total_orders (COUNT of order_id)

Only include customers who have placed more than 1 order.

Sort the results by total_orders in descending order.

Limit the results to the top 20 customers.

Include the following columns:
- customer_id
- first_name
- last_name
- total_orders

```python
import sqlite3

conn = sqlite3.connect("../data/bike_store.db")
curr = conn.cursor()

curr.execute("""
    SELECT c.customer_id, c.first_name, c.last_name, COUNT(o.order_id) AS total_orders
    FROM customers c LEFT JOIN orders o
    ON c.customer_id = o.customer_id
    WHERE c.state IN ("TX", "CA", "NY") AND
    c.last_name LIKE "%s%" OR
    c.city = "Seattle"
    GROUP BY c.customer_id, c.first_name, c.last_name
    HAVING COUNT(o.order_id) > 1
    ORDER BY total_orders DESC
    LIMIT 20
""").fetchall()
```
