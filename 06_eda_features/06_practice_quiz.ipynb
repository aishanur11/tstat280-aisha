### Practice Quiz

#### Part 1

Load the dataset found in `amazon_laptop_cleaned.csv`. Then create a visualization that would easily allow you to answer the following question:

- Which colors appear to be the most popular?

```python
import pandas as pd
import seaborn as sns

df = pd.read_csv("../data/amazon_laptop_cleaned.csv")

sns.countplot(data=df, x="colour_new")
```

#### Part 2

Create a visualization that would allow you to easily answer the following question:

- What does the distribution of the `price` of a laptop look like?

```python
sns.histplot(data=df, x="price")
```

#### Part 3

Create a visualization that would allow you to easily answer the following question:

- How does the distribution of `rating` differ for the following categories in `brand`: Lenovo, ASUS, and HP?

HINT: It might be helpful to subset the data before you make the visualization

```python
brand_subset = df[df["brand"].isin(["Lenovo", "ASUS", "HP"])]
sns.histplot(data=brand_subset, x="rating", hue="brand")
```

#### Part 4

Create a visualization that would allow you to easily answer the following question:

- What does the relationship between the log of `price` and `offer` look like?

```python
import numpy as np

df["price_log"] = np.log(df["price"])

sns.scatterplot(data=df, x="offer", y="price_log")
```

#### Part 5

Create a visualization that would allow you to easily answer the following question:

- Does there appear to any association between `brand` and the color of the laptop?

```python
colorbrand_crosstab = pd.crosstab(df["brand"], df["colour_new"])

sns.heatmap(colorbrand_crosstab, annot=True)
```

Think of any new feature that you would be interested in creating to see if it may have an association with `price`. Then create a visualization to confirm whether or not this new feature seems to be associated with `price`.
