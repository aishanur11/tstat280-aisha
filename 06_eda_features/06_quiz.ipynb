### Quiz

#### Part 1

Load the dataset found in `hr.csv`. Then create a visualization that would easily allow you to answer the following question:

- Which performance score is most common?

```python
import pandas as pd
import seaborn as sns

hr = pd.read_csv("../data//hr.csv")

sns.countplot(data=hr, x="PerformanceScore")
```

#### Part 2

Create a visualization that would allow you to easily answer the following question:

- What does the distribution of the `Salary` of an employee look like?

```python
sns.histplot(data=hr, x="Salary")
```

#### Part 3

Create a visualization that would allow you to easily answer the following question:

- How does the distribution of `EngagementSurvey` differ for the categories in `PerformanceScore`?

```python
sns.boxplot(data=hr, x="EngagementSurvey", hue="PerformanceScore")
```

#### Part 4

Create a visualization that would allow you to easily answer the following question:

- What does the relationship between the log of `Salary` and `Absences` look like?

```python
import numpy as np
hr["Salary_log"] = np.log(hr["Salary"])

sns.scatterplot(data=hr, x="Salary_log", y="Absences")
```

#### Part 5

Create a visualization that would allow you to easily answer the following question:

- Does there appear to any association between `EmploymentStatus` and `Department`

```python
empdep_crosstab = pd.crosstab(hr["EmploymentStatus"], hr["Department"])

sns.heatmap(empdep_crosstab, annot=True)
```
