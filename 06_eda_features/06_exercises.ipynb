#### Exercise 8

Perform a log transformation on `Salary` and then standardize the transformed values.

```python
import numpy as np
hr["Salary_log"] = np.log(hr["Salary"])

hr["Salary_transformed"] = hr["Salary_log"].pipe(lambda x: (x-x.mean())/(x.std()))
```

Now make a visualization to view the distribution of this new transformed and standardized variable.

```python
sns.histplot(data=hr, x="Salary_transformed")
```
