#### Part 1

Read in the dataset in `math_education.csv` and display the first 5 rows of the dataset.

```python
import pandas as pd
df = pd.read_csv("../data/math_education.csv")
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
df
```

#### Part 3

Treat any missing values for `study_hours` by replacing them with the median of `study_hours`.

```python
df["study_hours"] = df["study_hours"].fillna(df["study_hours"].median())
df["study_hours"].describe()
```

#### Part 4

Do the following with the `time_per_problem` variable:

- Clean up the text
- Convert to `float`
- Convert to everything to seconds

```python
MIN_TO_SEC = 60

time_multiplier = df["time_per_problem"].str.contains("minutes").map(
    lambda x: MIN_TO_SEC if x else 1
)

df["time_per_problem"] = df["time_per_problem"].str.replace("minutes", "").str.replace("seconds", "").astype("float")*time_multiplier


df["time_per_problem"].describe()                                                                                 
                                                                                 
```

#### Part 5

Remove any unwanted text ("%" and "percent") from the `test_score` variable and then convert this to a `float`.

```python
df["test_score"] = df["test_score"].str.replace("%", "").str.replace("percent", "").astype("float")

df["test_score"].describe()
```

#### Part 6

Standardize the categories in `learning_style` using the following mapping:

- Audio, auditory, AUDITORY, Auditory → Auditory
- kinesthetic, KINESTHETIC, Kinesthetic (with space), Kinesthetic → Kinesthetic
- visual, VISUAL, Visual → Visual

(Note that one of the values for the Kinesthetic category has one extra space that needs to be removed)

```python
df["learning_style"].unique()
```

```python
df["learning_style"] = df["learning_style"].str.replace(" ", "")
```

```python
df["learning_style"].unique()
```

```python
learning_dict = {}

learning_dict["Audio"] = "Auditory"
learning_dict["auditory"] = "Auditory"
learning_dict["AUDITORY,"] = "Auditory"
learning_dict["Auditory"] = "Auditory"

learning_dict["kinesthetic"] = "Kinesthetic"
learning_dict["KINESTHETIC,"] = "Kinesthetic"
learning_dict["Kinesthetic"] = "Kinesthetic"

learning_dict["visual"] = "Visual"
learning_dict["VISUAL"] = "Visual"
learning_dict["Visual"] = "Visual"


df["learning_style"] = df["learning_style"].map(learning_dict)

df["learning_style"].describe()
```
