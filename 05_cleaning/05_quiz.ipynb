### Quiz

#### Part 1

Read in the dataset in `nutrition.csv` and display the first 5 rows of the dataset.

```python
import pandas as pd
df = pd.read_csv("../data/nutrition.csv")
df.head()
```

#### Part 2

Drop any duplicate rows (rows where every value is the same).

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
df.info()
```

#### Part 3

Treat any missing values for `calories` by replacing them with the mean of `calories`.

```python
df["calories"].isnull().sum()
```

```python
df["calories"] = df["calories"].fillna(df["calories"].mean())

df["calories"].describe()
```

#### Part 4

Do the following with the `protein` variable:

- Clean up the text
- Convert to `float`
- Convert to everything to ounces (1 oz is 28.3495 grams)

```python
OZ_TO_GRAMS = 28.3495

protein_multiplier = df["protein"].str.contains("oz").map(
    lambda x: OZ_TO_GRAMS if x else 1
)

df["protein"] = df["protein"].str.replace("g", "").str.replace("oz", "").astype("float")*protein_multiplier

df["protein"].describe()
```

```python
df["BMI"] = df["BMI"].str.replace("BMI", "").str.replace("bmi", "").astype("float")

df["BMI"].describe()
```

#### Part 6

Standardize the categories in `diet` using the following mapping:

- Omni, omnivore → Omnivore
- other, OtherDiet → Other
- vegan, VGN → Vegan
- vegetarian, Veggie → Vegetarian

```python
diet_dict={}

diet_dict["Omni"] = "Omnivore"
diet_dict["omnivore"] = "Omnivore"

diet_dict["other"] = "Other"
diet_dict["OtherDiet"] = "Other"

diet_dict["vegan"] = "Vegan"
diet_dict["VGN"] = "Vegan"

diet_dict["vegetarian"] = "Vegetarian"
diet_dict["Veggie"] = "Vegetarian"


df["diet"] = df["diet"].map(diet_dict)

df["diet"].describe()
```
