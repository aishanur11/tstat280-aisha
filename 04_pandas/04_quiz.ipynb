#### Part 3

Create a new column called `box_volume` that is the product of dimensions `x`, `y`, and `z`. Then display the new DataFrame.

```python
df["box_volume"] = df["x"] * df["y"] * df["z"]
df
```

#### Part 4

Use either `map` or `apply` (whichever is more appropriate) to create a new column called `good_deal` which is 1 if the cut is Ideal and the price is below 500. Then display the new DataFrame.

```python
df["good_deal"] = df.apply(
    lambda row: 1 if (row["cut"] == "Ideal" and row["price"] < 500 )
                    else 0, axis=1
)
df
```

#### Part 6

Merge the `movies` and `ratings` DataFrames so that all information about the movies in `movies` is kept even if information about their rating is not available in `ratings`. Display the merged DataFrame.

```python
movies = pd.DataFrame({
    "movie_id": [10, 11, 12, 13],
    "title": ["Inception", "Titanic", "Avatar", "Jaws"]
})

ratings = pd.DataFrame({
    "movie_id": [10, 11, 14],
    "rating": [8.8, 7.8, 6.5]
})

```

```python
movies.merge(ratings, on="movie_id", how="left")
```
