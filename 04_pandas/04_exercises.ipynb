#### Exercise 7

1. Use the `map` method to create a new column named `is_expensive` that is 1 if the order cost is greater than 30 and is 0 otherwise.
2. Use the `apply` method to create a new column named `fast_favorite` which is 1 if the cuisine type is `"Japanese"` and the cost is below 20 dollars and is 0 otherwise


```python
df["is_expensive"] = df["order_cost"].map(
    lambda x: 1 if x > 30 
    else 0
)
df.head()
```

```python
df["fast_favorite"] = df.apply(
    lambda row: 1 if (row["cuisine_type"] == "Japanese" and row["order_cost"] < 20)
                 else 0, axis = 1
)
df.head()
```

#### Exercise 11

Using the Pandas DataFrames provided to you, complete the following:

1. Merge `df_orders` and `df_customers` on `customer_id` using an inner join.
2. How many rows are in the resulting DataFrame?
3. Perform a left join using `df_orders` as the left DataFrame. What differences do you observe compared to the inner join?

```python
df_orders = pd.DataFrame({
    'order_id': [1, 2, 3, 4],
    'customer_id': [101, 102, 103, 104],
    'cost': [25, 40, 15, 30]
})

df_customers = pd.DataFrame({
    'customer_id': [101, 102, 103, 105],
    'customer_name': ['Alice', 'Bob', 'Charlie', 'Diana']
})

df_inner = df_customers.merge(df_orders,on="customer_id")
print(df_inner)
```

```python
df_left = df_orders.merge(df_customers, on="customer_id", how="left")
print(df_left)
```

#### Exercise 12

Using the Pandas DataFrames provided to you, complete the following:

1. Merge the two DataFrames where `restaurant_id` matches `rest_id`.
2. Perform an outer join and examine the resulting DataFrame.
3. Identify which rows contain missing values after the merge and explain why they appear.

```python
df_restaurants = pd.DataFrame({
    'restaurant_id': [1, 2, 3],
    'restaurant_name': ['Pizza Place', 'Sushi Spot', 'Burger Joint']
})

df_reviews = pd.DataFrame({
    'rest_id': [1, 1, 2, 4],
    'rating': [4.5, 4.0, 5.0, 3.5]
})

df_match = df_restaurants.merge(df_reviews, left_on="restaurant_id", right_on="rest_id")
print(df_match)
```

```python
df_outer = df_restaurants.merge(df_reviews, left_on="restaurant_id", right_on="rest_id", how="outer")
print(df_outer)
```
