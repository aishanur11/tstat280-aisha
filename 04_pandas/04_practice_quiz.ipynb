#### Part 3

Create a new column called `price_per_carat` that divides `price` by `carat`. Then output the new DataFrame.

```python
df["price_per_carat"] = df["price"] / df["carat"]
df
```

#### Part 4

Use either `apply` or `map` (whichever is more appropriate) to create a new column called `is_expensive` that is 1 if price is greater than 5000 and is 0 otherwise. Then output the new DataFrame.

```python
df["is_expensive"] = df["price"].map(
    lambda x: 1 if x > 5000
    else 0
)
df
```

#### Part 5

```python
df.groupby("cut")["price"].mean()
```

#### Part 6

Merge the `students` and `majors` DataFrames so that all information about the students in `students` is kept even if information about their major is not available in `majors`. Output the merged DataFrame.

```python
students = pd.DataFrame({
    "student_id": [101, 102, 103, 104],
    "name": ["Alice", "Bob", "Charlie", "Diana"]
})

majors = pd.DataFrame({
    "student_id": [101, 102, 105],
    "major": ["Biology", "Computer Science", "Math"]
})
df_left = students.merge(majors, on="student_id", how="left")
print(df_left)
```
